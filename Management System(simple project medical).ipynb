```python
import tkinter as tk
from tkinter import simpledialog,messagebox
class Patient:
    def __init__(self,patient_id,name,phone_number,birth_day,location,visit_date):
        self.patient_id=patient_id
        self.name=name
        self.phone_number=phone_number
        self.birth_day=birth_day
        self.location=location
        self.visit_date=visit_date

    def update_info(self,name,phone_number,birth_day,location,visit_date):
        if name:
            self.name=name
        
        if phone_number:
            self.phone_number=phone_number

        if birth_day:
            self.birth_day

        if location:
            self.location=location
        
        if visit_date:
            self.visit_date=visit_date

    def __str__(self):    #dunder function (polymorishm)
        return (f"ID: {self.patient_id}, Name: {self.name}, Phone: {self.phone_number}, Birthday Date: {self.birth_day}, Address: {self.location}, Visit Date: {self.visit_date}.")
    

```

```python
class Clinic:
    def __init__(self):
        self.patients={}    #dictionary take every patient

    def add_patient(self,newpatient):
        self.patients[newpatient.patient_id]=newpatient

    def remove_patient(self,patient_id):
        if patient_id in self.patients:
            del self.patients[patient_id]

    def get_patient(self,patient_id):
        return self.patients.get(patient_id)
    
    def list_patient(self):
        return [str(patient) for patient in self.patients.values()]   #List Comprehension
```

```python


class ClinicGui:
    def __init__(self,root):
        self.clinic=Clinic()   #Clinc()   ->class
        self.root=root
        self.root.title("Clinic Management System")
        self.setup_gui()

        self.root.configure(bg="lightblue") 

    def setup_gui(self):

        tk.Label(self.root,text="Patient ID:",bg="lightblue").grid(row=0,column=0,padx=10,pady=5)
        self.patient_id_entry=tk.Entry(self.root)
        self.patient_id_entry.grid(row=0, column=1, padx=10, pady=5)

        tk.Label(self.root, text="Name:",bg="lightblue").grid(row=1, column=0, padx=10, pady=5)
        self.name_entry = tk.Entry(self.root)
        self.name_entry.grid(row=1, column=1, padx=10, pady=5)

        tk.Label(self.root, text="Phone Number:",bg="lightblue").grid(row=2, column=0, padx=10, pady=5)
        self.phone_number_entry = tk.Entry(self.root)
        self.phone_number_entry.grid(row=2, column=1, padx=10, pady=5)

        tk.Label(self.root, text="Date of Birth (YYYY-MM-DD):",bg="lightblue").grid(row=3, column=0, padx=10, pady=5)
        self.birth_day_entry = tk.Entry(self.root)
        self.birth_day_entry.grid(row=3, column=1, padx=10, pady=5)

        tk.Label(self.root, text="Location:",bg="lightblue").grid(row=4, column=0, padx=10, pady=5)
        self.location_entry = tk.Entry(self.root)
        self.location_entry.grid(row=4, column=1, padx=10, pady=5)

        tk.Label(self.root, text="Visit Date (YYYY-MM-DD):",bg="lightblue").grid(row=5, column=0, padx=10, pady=5)
        self.visit_date_entry = tk.Entry(self.root)
        self.visit_date_entry.grid(row=5, column=1, padx=10, pady=5)



        tk.Button(self.root,text="Add patient",command=self.add_patient,bg='white',fg="blue").grid(row=6, column=0, columnspan=1, pady=10)
        tk.Button(self.root,text="Rmove patient",command=self.remove_patient,bg='white',fg="blue").grid(row=6, column=1, columnspan=1, pady=10)
        tk.Button(self.root,text="list patients",command=self.list_patient,bg='white',fg="blue").grid(row=6, column=2, columnspan=1, pady=10)

    def add_patient(self):
        patient_id=self.patient_id_entry.get()
        Name=self.name_entry.get()
        Phone=self.phone_number_entry.get()
        Birthday=self.birth_day_entry.get()
        Location=self.location_entry.get()
        Visit_Date=self.visit_date_entry.get()

        if patient_id and Name and Phone and Birthday and Location and Visit_Date:
            patient=Patient(patient_id , Name , Phone , Birthday , Location , Visit_Date)
            self.clinic.add_patient(patient)
            messagebox.showinfo("Success", "Patient added successfully!")
            self.clear_entries()
        else:
            messagebox.showwarning("Input Error", "Please fill in all required fields.")

    
    def remove_patient(self):
        patient_id=self.patient_id_entry.get()
        if patient_id:
            self.clinic.remove_patient(patient_id)
            messagebox.showinfo("Success", "Patient removed successfully!")
            self.clear_entries()
        else:
            messagebox.showwarning("Input Error", "Please enter a valid Patient ID.")

    def list_patient(self):
        patients=self.clinic.list_patient()

        if patients:
            messagebox.showinfo("Patient List: ","\n".join(patients))
        else:
            messagebox.showwarning("Patient List", "No patients found.")

    def clear_entries(self):
        self.patient_id_entry.delete(0,tk.END)
        self.name_entry.delete(0,tk.END)
        self.phone_number_entry.delete(0,tk.END)
        self.birth_day_entry.delete(0,tk.END)
        self.location_entry.delete(0,tk.END)
        self.visit_date_entry.delete(0,tk.END)


if __name__== '__main__':
    root=tk.Tk()
    app=ClinicGui(root)
    root.mainloop()
       
```
